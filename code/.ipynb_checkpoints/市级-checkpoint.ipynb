```python
print(type(Map), Map.head(5))
```

```python
dm=Map['市代码'].tolist()
dm=[str(j)[:4] for j in dm]
```

```python
testdata=pd.read_csv('../data/invest_after.csv')
```

```python
company=testdata[['company_name','company_industry_code','company_area_code']]
invested_company=testdata[['invested_company_name','invested_company_industry_code','invested_company_area_code']]
company.columns=['company','industry_code','area_code']
invested_company.columns=['company','industry_code','area_code']
all_com=pd.concat([company,invested_company])
all_com=all_com.drop_duplicates(subset=['company']).reset_index()
lst=all_com['industry_code'].tolist()
lst=[i[:1] for i in lst]
all_com['industry_code']=lst
all_com
```

# A

```python
coma=all_com[all_com['industry_code']=='A']
```

```python

area=coma['area_code'].tolist()
area=[str(j) for j in area]
for i in range(len(area)):
    area[i]=area[i][:4]
# print(area)

num_city=[]
for i in dm:
    num_city.append(area.count(i))
```

```python
dm=[int(j)*100 for j in dm]
```

```python
Data=pd.DataFrame({'code':dm,'num':num_city})
Data
```

```python
Data_with_Map = pd.merge(left=Map, right=Data, how='left',
                         left_on='市代码', right_on='code')
# Data_with_Map.plot(column='num', cmap='OrRd', legend=True,
#                    missing_kwds={
#                        "color": "grey",
#                        "edgecolor": "black",
#                        "hatch": ".",
#                        "alpha": 0.2
#                    })
# plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))           # 图设置
ax = fig.add_axes((0.05, 0.05, 0.9, 0.95))  # 轴设置

divider = make_axes_locatable(ax)                         # 保证colorbar不过长
cax = divider.append_axes("right", size="5%", pad=0.1)

albers_proj = '+proj=aea +lat_1=25 +lat_2=47 +lon_0=105'  # 设置CRS

# 绘图 -- 大中国地图
ax = Data_with_Map.to_crs(albers_proj).plot(ax=ax,    # 轴
                                            cax=cax,  # colorbar轴
                                            column='num',
                                            cmap='OrRd',
                                            edgecolor='black',  # 线的颜色
                                            linewidth=0.3,      # 线宽
                                            linestyle='--',     # 线形
                                            missing_kwds={
                                                "color": "lightgrey",
                                                "edgecolor": "black",
                                                "hatch": ".",
                                                "alpha": 0.2
                                            },
                                            legend=True)
                                            
# 创建子图前的设置
bound = gpd.GeoDataFrame({          # 设置经纬坐标范围
    'x': [80, 150, 106.5, 123],
    'y': [15, 50, 2.8, 24.5]
})

bound.geometry = bound.apply(lambda row: Point(
    [row['x'], row['y']]), axis=1)  # 添加矢量列
bound.crs = 'EPSG:4326'  # 初始化CRS
bound.to_crs(albers_proj, inplace=True)  # 再投影
bound

# 设置图坐标范围
ax.set_xlim(bound.geometry[0].x, bound.geometry[1].x)
ax.set_ylim(bound.geometry[0].y, bound.geometry[1].y)

ax.set_xticks([])
ax.set_yticks([])

# 创建南海插图对应的子图
ax_child = fig.add_axes([0.7, 0.15, 0.2, 0.2])  # 设置子图位置
ax_child = Data_with_Map.to_crs(albers_proj).plot(ax=ax_child,   # 子图的绘制 -- 除了轴用自己的，其余照搬
                                                  column='num',
                                                  cmap='OrRd',
                                                  edgecolor='black',  
                                                  linewidth=0.3,      
                                                  linestyle='--',     
                                                  missing_kwds={
                                                      "color": "lightgrey",
                                                      "edgecolor": "black",
                                                      "hatch": ".",
                                                      "alpha": 0.2
                                                  })

ax_child = Line.geometry.to_crs(albers_proj).plot(ax=ax_child,  # 绘制九段线
                                                  edgecolor='grey',
                                                  linewidth=1,
                                                  alpha=0.4)
                                                  
# 设置子图范围 -- 仅展示中国的右下角
ax_child.set_xlim(bound.geometry[2].x, bound.geometry[3].x)
ax_child.set_ylim(bound.geometry[2].y, bound.geometry[3].y)

# 设置子图坐标 -- 无坐标
ax_child.set_xticks([])
ax_child.set_yticks([])



plt.show()
```
