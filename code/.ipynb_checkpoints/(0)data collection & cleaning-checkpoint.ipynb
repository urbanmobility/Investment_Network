```python
import numpy as np
import pandas as pd
import os
import requests
from bs4 import BeautifulSoup
import json
```

# lng & lat

```python
# original data
invest = pd.read_csv('../data/invest.csv')
invest
```

```python

invest_data = invest[(~invest['company_registered_address'].isnull()) & (~invest['invested_company_registered_address'].isnull())]
invest_data = invest_data.reset_index()
invest_data
```

```python
len(invest_data)
```

```python
result=[]
```

```python
#for i in range(300000, len(invest_data)):
for i in range(10):
    add1 = invest_data.loc[i]['company_registered_address']
    add2 = invest_data.loc[i]['invested_company_registered_address']
    url1 = 'http://api.map.baidu.com/geocoding/v3/?address=%s&output=json&ak=dOj0PS8c71nXcYsABFvB01gEdpDEgV54&callback=showLocation&ret_coordtype=WGS84' % add1
    url2 = 'http://api.map.baidu.com/geocoding/v3/?address=%s&output=json&ak=dOj0PS8c71nXcYsABFvB01gEdpDEgV54&callback=showLocation&ret_coordtype=WGS84' % add2
    try:
        html1 = requests.get(url1) 
        html2 = requests.get(url2)
        text1 = json.loads(html1.text.replace('showLocation&&showLocation(', '')[:-1])
        text2 = json.loads(html2.text.replace('showLocation&&showLocation(', '')[:-1])
        if text1['status'] == 0 and text2['status'] == 0:
#             print(i)
            print(text1)
            result.append([invest_data.loc[i]['company_name'],
                           text1['result']['location']['lng'],
                           text1['result']['location']['lat'],
                           invest_data.loc[i]['invested_company_name'],
                           text2['result']['location']['lng'],
                           text2['result']['location']['lat']])
    except:
        continue
```

```python
result
```

```python
invest_data_address = pd.DataFrame(result, columns=['company_name','company_lng','company_lat','invested_company_name','invested_company_lng','invested_company_lat'])
invest_data_address
```

```python
invest_data_address = pd.merge(invest_data_address,invest_data,how='left')
invest_data_address
```

```python
# invest_data_address
invest_data_address['invested_company_area_code'] = invest_data_address['invested_company_area_code'].astype(str)
invest_data_address['company_area_code'] = invest_data_address['company_area_code'].astype(str)
```

```python
invest_data_address = invest_data_address.drop_duplicates(subset=['company_name','invested_company_name'])
```
