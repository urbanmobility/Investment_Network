```python
# dOj0PS8c71nXcYsABFvB01gEdpDEgV54
def reverse_geocoding(lng, lat, currentkey):
    """
    lat and lng convert address
    :param lng: longitude
    :param lat: latitude
    :param currentkey: AK
    :return: places_ll
    """
    url = 'http://api.map.baidu.com/reverse_geocoding/v3/?'
    params = {
        "location": str(lat)+','+str(lng),
        "output": 'json',
        "ak": currentkey,
        "coordtype": "wgs84ll",
    }
    response = requests.get(url, params=params)
    answer = response.json()
    if answer['status'] == 0:
        tmpList = answer['result']
        address = tmpList['formatted_address']
        if tmpList['addressComponent']['province']!=tmpList['addressComponent']['city']:
            district=tmpList['addressComponent']['province']+tmpList['addressComponent']['city']+tmpList['addressComponent']['district']
        else:
            district=tmpList['addressComponent']['province']+tmpList['addressComponent']['district']
        areacode=tmpList['addressComponent']['adcode']
        print([district,areacode])
        return district,areacode

    else:
        return -1
```

```python
def geocoding(address, currentkey):
    """
    address convert lat and lng
    :param address: address
    :param currentkey: AK
    :return: places_ll
    """
    url = 'http://api.map.baidu.com/geocoding/v3/?'
    params = {
        "address": address,
        "output": 'json',
        "ak": currentkey,
    }
    response = requests.get(url, params=params)
    answer = response.json()
    if answer['status'] == 0:
        tmpList = answer['result']
        coordString = tmpList['location']
        coordList = [coordString['lng'], coordString['lat']]
        district_lng,district_lat=float(coordList[0]), float(coordList[1])
#         print([district_lng,district_lat])
        return district_lng,district_lat
    
    else:
        district_lng=0
        district_lat=0
        return district_lng,district_lat
```

# power law distribution

```python
comlst=set(testdata['company_area_code'].tolist()+testdata['invested_company_area_code'].tolist())
comlst=list(comlst)

degree=[]
for i in tqdm(comlst):
    data1=testdata[testdata['company_area_code']==i]['invested_company_area_code'].tolist()
    data2=testdata[testdata['invested_company_area_code']==i]['company_area_code'].tolist()
    data=set(data1+data2)
    degree.append(len(data))
degree
```

```python
import matplotlib.pyplot as plt

plt.rc('font',family='Arial')

degree_x=tuple(set(degree))
degree_y=tuple([degree.count(i)/len(degree) for i in degree_x])
# degree_x,degree_y

plt.figure(figsize=[6,5])
plt.ylabel('Frequency',fontsize=16)
plt.xlabel('Degree',fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.scatter(degree_x,degree_y,color=['#d73027'])

bwith = 2
ax = plt.gca()
ax.spines['bottom'].set_linewidth(bwith)
ax.spines['left'].set_linewidth(bwith)
ax.spines['top'].set_linewidth(bwith)
ax.spines['right'].set_linewidth(bwith)

plt.show()
```

```python
import math
from scipy import optimize

degree_x_ = np.array([math.log(i) for i in degree_x]) 
degree_y_ = np.array([math.log(i) for i in degree_y])
plt.figure(figsize=[6,5])
# plt.xscale('log')
# plt.yscale('log') 
# plt.title('')
plt.ylabel('Log of frequency',fontsize=16)
plt.xlabel('Log of degree',fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.scatter(degree_x_,degree_y_,color=['#4575b4'])
plt.show()
```

```python
from sklearn.metrics import r2_score
plt.figure(figsize=[6,5])
plt.scatter(degree_x_,degree_y_,color=['#d73027'])

x=[i for i in degree_x_ if i<=5]
y=[degree_y_[i] for i in range(len(degree_y_)) if degree_x_[i]<=5]
# x=[i for i in degree_x_]
# y=[degree_y_[i] for i in range(len(degree_y_))]
slope, intercept = np.polyfit(x, y, 1)

pred=[]
for i in x:
    pred.append(slope * i + intercept)
r2=r2_score(y, pred)

# x.append(4.3)
x=np.array(x)
plt.plot(x, slope * x + intercept, color='black',linestyle='dashed',linewidth=2)
plt.ylabel('Log of frequency',fontsize=16)
plt.xlabel('Log of degree',fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

bwith = 2 
ax = plt.gca()
ax.spines['bottom'].set_linewidth(bwith)
ax.spines['left'].set_linewidth(bwith)
ax.spines['top'].set_linewidth(bwith)
ax.spines['right'].set_linewidth(bwith)

plt.text(x=1,y=-1, s='log(Y) = %.2f*log(X) +%.2f, log(X)<=5\nr$^2$ = %.2f'%(slope,intercept,r2), fontsize=16)
plt.show()
```
