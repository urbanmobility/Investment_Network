```python
import numpy as np
import pandas as pd
import os
import requests
from bs4 import BeautifulSoup
import json
```

```python
testdata=pd.read_csv('../data/invest_cleaned.csv')
testdata
```

```python
hy=['A','B','C','D','E','F','G','H','I','J','K','L','M','N','O','P','Q','R']
```

# 27+4 cities

```python
dm=[11,12,31,50,1301,1401,1501,2101,2201,2301,3201,3301,3401,3501,3601,3701,4101,4201,4301,4401,
          4501,4601,5101,5201,5301,5401,6101,6201,6301,6401,6501]
dm=[str(j) for j in dm]
```

```python
lst1=list(testdata['company_area_code'].astype(str))
for i in dm:
    a=len(i)
    for j in range(len(lst1)):
        if lst1[j][:a] == i:
            lst1[j]=i

lst2=list(testdata['invested_company_area_code'].astype(str))
for i in dm:
    a=len(i)
    for j in range(len(lst2)):
        if lst2[j][:a] == i:
            lst2[j]=i

all_sim=pd.DataFrame({'company_area_code':lst1,'invested_company_area_code':lst2})

indus1=testdata['company_industry_code'].tolist()
indus1=[str(j) for j in indus1]
for i in range(len(indus1)):
    indus1[i]=indus1[i][:1]
    
indus2=testdata['invested_company_industry_code'].tolist()
indus2=[str(j) for j in indus2]
for i in range(len(indus2)):
    indus2[i]=indus2[i][:1]

all_sim_n=pd.DataFrame({'company_industry_code':indus1,'invested_company_industry_code':indus2})
all_sim_n=pd.concat([all_sim_n,all_sim],axis=1)
all_sim_n
```

```python
# outward investment

I_out=[]
for i in dm:
    num=[]
    lst=all_sim_n[all_sim_n['company_area_code']==i]
    for j in hy:
        num.append(list(lst['company_industry_code']).count(j))
    I_out.append(num)
print(I_out)
```

```python
# inward investment

I_in=[]
for i in dm:
    num=[]
    lst=all_sim_n[all_sim_n['invested_company_area_code']==i]
    for j in hy:
        num.append(list(lst['invested_company_industry_code']).count(j))
    I_in.append(num)
print(I_in)
```
